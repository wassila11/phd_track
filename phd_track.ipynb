```python
import pandas as pd
```

```python
# read csv database and display it
food_db = pd.read_csv("Table Ciqual 2020_ENG_2020 07 07.csv")
food_db
```

```python
# Nettoyage de la base de données
# 1. Suppression de l;a premiere ligne qui contient que des valeurs nules
df = food_db.copy()
df_cleaned = df.iloc[1:].reset_index(drop=True)

# 2. Suppression des repas composés
df_cleaned = df_cleaned[~df_cleaned["alim_grp_nom_eng"].isin(["starters and dishes", "beverages"])].copy()
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned["alim_grp_nom_eng"].unique()
```

```python
# 3. Suppression des colonnes inutiles
df_cleaned.drop(
    [
        "alim_ssgrp_code",
        "alim_grp_code",
        "alim_ssssgrp_code",
        "alim_code",
        "alim_nom_sci", 
    ],
    axis=1,
    inplace=True,
)
```

```python
# Renommage des colonnes
df_cleaned.rename(
    columns={
    'alim_grp_nom_eng': 'Main category',
    'alim_ssgrp_nom_eng': 'Subcategory',
    'alim_ssssgrp_nom_eng': 'Detailed category',
    'alim_nom_eng': 'Ingredient',
    'Energy, Regulation EU No 1169/2011 (kJ/100g)': 'Energy_kj',
    'Energy, Regulation EU No 1169/2011 (kcal/100g)': 'Energy_kcal',
    },
    inplace=True,
)

df_cleaned.columns
```

```python
# Convertir toutes les colonnes nutritionnelles en float (sauf la colonne du nom d'ingrédient)
for col in df_cleaned.columns:
    if col != "Main category" and col != "Subcategory" and col != "Detailed category" and col != "Ingredient":
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# Remplacer les valeurs avec '<' par des approximations, et '-' par NaN
df_cleaned = df_cleaned.applymap(
    lambda x: 0.05 if isinstance(x, str) and x.strip().startswith('<') else
              None if x == '-' else x
)
df_cleaned.fillna(0, inplace=True)
```

```python
# Save the cleaned dataframe to a new csv file
df_cleaned.to_csv("food_db_cleaned.csv", index=False)
```
